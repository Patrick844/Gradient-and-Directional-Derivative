```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter

```

```python
def function(x,y):
    return (x**3)*y - (x*y**3)
```

```python
def derivative_X(x,y):
    return 3*x**2*y-y**3
    
```

```python
def derivative_Y(x,y):
    return x**3-3*x*y**2
    
```

```python
def normalize_u(u):
    length_u=np.linalg.norm(u)
    normalized_u= u / length_u
    return normalized_u
    
    
```

```python
def directional_derivative(normalized_u,gradient):
    return gradient@normalized_u
```
